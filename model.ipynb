```python
import json
```

```python
# Function to create JSONL lines
def create_jsonl_line(row):
    return json.dumps({
        'custom_id': f'custom_id_{row["MAL_ID"]}',
        'method': 'POST',
        'url': '/v1/embeddings',
        'body': {
            'model': 'text-embedding-3-small',
            'input': row['content'],
            'encoding_format': 'float'
        }
    })
```

```python
# Write to JSONL file
with open('batch_requests.jsonl', 'w') as f:
    for _, row in df.iterrows():
        f.write(create_jsonl_line(row) + '\n')
```

```python
# Upload the file
response = client.files.create(
    file=open('batch_requests.jsonl', 'rb'),
    purpose='batch'
)
file_id = response.id
```

```python
client.batches.create(
    input_file_id=file_id,
    endpoint="/v1/embeddings",
    completion_window="24h",
)
```

```python
result_file_id = batch.output_file_id
result = client.files.content(result_file_id).content
```

```python
# Loading data from saved file
results = []
with open(result_file_name, 'r') as file:
    for line in file:
        # Parsing the JSON string into a dict and appending to the list of results
        json_object = json.loads(line.strip())
        results.append(json_object)
```

```python
import pandas as pd
import numpy as np
import faiss

# Convert the embedding column to a NumPy array
embeddings = np.array(df["embedding"].to_list(), dtype=np.float32)

# Step 1: Build the FAISS index
dimension = len(embeddings[0])  # Dimension of the embeddings
index = faiss.IndexFlatL2(dimension)  # L2 distance index (Euclidean)

# Add embeddings to the FAISS index
index.add(embeddings)

# Step 2: Define the query vector
# Example: Use the embedding of the first row as the query
query_vector = np.array([embeddings[110]], dtype=np.float32)

# Step 3: Perform a k-nearest neighbors search
k = 5  # Number of nearest neighbors
distances, indices = index.search(query_vector, k)

# Step 4: Print the results
print("Query vector:", query_vector)
print("Nearest neighbors' indices:", indices)
print("Nearest neighbors' distances:", distances)

# Retrieve the MAL_ID and content for the results
for idx in indices[0]:
    print("MAL_ID:", df.iloc[idx]["MAL_ID"])
    print("Content:", df.iloc[idx]["content"])
```
